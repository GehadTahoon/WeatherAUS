```python
# constants
OBJ_COLS = ['Location','WindGustDir','WindDir9am','WindDir3pm','RainToday']

EXPORT_PATH = 'data/preprocessed.pkl'
```

```python
# import packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()

import pickle

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler


from sklearn.impute import KNNImputer
```

```python
raw_data = pd.read_csv('data/Weather Training Data.csv')
raw_data.drop('row ID',axis=1,inplace=True)
```

### _Encode categorical data_


```python
lb = LabelEncoder()
encoded_data = raw_data.copy()
for col in OBJ_COLS:
    encoded_data[col] = lb.fit_transform(raw_data[col])
```

```python
# Replacing missing values with median of them
null_cols = encoded_data.drop(['Location','RainTomorrow'],axis=1).columns
processed_data = encoded_data.copy()

processed_data[null_cols] = processed_data[null_cols].fillna(processed_data[null_cols].median())
```

### _Data scaling_

```python
scaler = MinMaxScaler()
scale_cols = processed_data.drop('RainTomorrow',axis=1).columns
data = pd.DataFrame( scaler.fit_transform(processed_data[scale_cols]),columns=scale_cols)
data = pd.concat([data,processed_data['RainTomorrow']],axis=1)
```
